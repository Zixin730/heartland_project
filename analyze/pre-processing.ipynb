```python
import pandas as pd
import numpy as np
```

# Get all tickers

```python
members = pd.read_excel("../data/members.xlsx")
members
```

```python
members_arr = np.array(members)[:, 1:]
members_arr
```

```python
members_ls = members_arr.reshape((members_arr.shape[0]*members_arr.shape[1], ))
```

```python
members_set = list(set(members_ls))
members_set = members_set[1:]
file_name = "../data/members.txt"
f = open(file_name, 'w') 
for name in members_set:
    f.write(name+"\n")
f.close()
```
